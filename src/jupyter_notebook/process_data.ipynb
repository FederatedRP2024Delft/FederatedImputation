```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
# Performance of VAE
```

```python
clients_data = []
for dirichlet in np.linspace(0.1,1, 10):
    dirichlet = round(dirichlet, 1)
    path = f"C:\\Users\\lyada\\Desktop\\Federated-Learning-PyTorch\\TheArchive\\objects\\{dirichlet}_mnist_vae_30_C[1.0]_iid[2]_E[10]_B[10].pkl"
    with open(path, 'rb') as f:
        clients_data.append(pickle.load(f))
    plt.plot(np.mean(clients_data[-1][1], axis=0), label = f"vae_{dirichlet}")

plt.title("Average Test Loss vs Global Round on MNIST")
plt.xlabel('Global Round')
plt.ylabel('Test Loss')

plt.legend()
plt.show()
```
